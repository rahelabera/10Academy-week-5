```python
import pandas as pd

credit_df = pd.read_csv('../data/raw/creditcard.csv')

# Check missing values
print(credit_df.isnull().sum())  # Typically none

# Remove duplicates
credit_df.drop_duplicates(inplace=True)

# Data types are usually correct (Time float, V1-V28 float, Amount float, Class int)

# Save cleaned
credit_df.to_csv('../data/processed/cleaned_creditcard.csv', index=False)
```

```python
# Univariate
import seaborn as sns
import matplotlib.pyplot as plt
import os
os.makedirs('figures', exist_ok=True)
sns.histplot(credit_df['Amount'], bins=50, kde=True)
plt.title('Distribution of Transaction Amount')
plt.savefig('figures/amount_dist.png')

# Bivariate (harder due to anonymized features, but example with Amount)
sns.boxplot(x='Class', y='Amount', data=credit_df)
plt.title('Amount vs Fraud Class')
plt.savefig('figures/amount_vs_class.png')

# Class distribution
class_dist_credit = credit_df['Class'].value_counts(normalize=True) * 100
print(class_dist_credit)  # Typically extreme: 0: 99.8%, 1: 0.2%
sns.countplot(x='Class', data=credit_df)
plt.title('Class Distribution (Credit Card)')
plt.savefig('figures/class_dist_credit.png')
```
